```python
import requests
import pandas as pd
```

```python
ob_adlt_url = "https://ghoapi.azureedge.net/api/NCD_BMI_30C"

ob_chld_url = "https://ghoapi.azureedge.net/api/NCD_BMI_PLUS2C"

mal_adlt_url = "https://ghoapi.azureedge.net/api/NCD_BMI_18C"

mal_chld_url = "https://ghoapi.azureedge.net/api/NCD_BMI_MINUS2C"

ob_adlt_result = requests.get(ob_adlt_url)

ob_chld_result = requests.get(ob_chld_url)

mal_adlt_result = requests.get(mal_adlt_url)

mal_chld_result = requests.get(mal_chld_url)

ob_adlt_data = ob_adlt_result.json()

ob_chld_data = ob_chld_result.json()

mal_adlt_data = mal_adlt_result.json()

mal_chld_data = mal_chld_result.json()
```

```python
flt_col_ob = ob_df[['ParentLocation', 'Dim1', 'TimeDim', 'Low', 'High', 'NumericValue', 'SpatialDim' ]]
flt_col_mal = mal_df[['ParentLocation', 'Dim1', 'TimeDim', 'Low', 'High', 'NumericValue', 'SpatialDim' ]]
```

```python
flt_col_ob.rename(columns={'ParentLocation' : 'Region', 'Dim1' : 'Gender', 'TimeDim' : 'Year', 'Low' : 'LowerBound', 'High' : 'UpperBound', 'NumericValue' : 'Mean_Estimate', 'SpatialDim' : 'Country'}, inplace=True) 
flt_col_mal.rename(columns={'ParentLocation' : 'Region', 'Dim1' : 'Gender', 'TimeDim' : 'Year', 'Low' : 'LowerBound', 'High' : 'UpperBound', 'NumericValue' : 'Mean_Estimate', 'SpatialDim' : 'Country'}, inplace=True)
```

```python
row_ob = flt_col_ob[(flt_col_ob['Year'] >=2012) & (flt_col_ob['Year'] <=2022)]
row_mal = flt_col_mal[(flt_col_mal['Year'] >=2012) & (flt_col_mal['Year'] <=2022)]
```

```python
row_ob['Gender'] = row_ob['Gender'].map({'SEX_MLE' : 'male', 'SEX_FMLE' : 'female', 'SEX_BTSX' : 'both'})
row_mal['Gender'] = row_mal['Gender'].map({'SEX_MLE' : 'male', 'SEX_FMLE' : 'female', 'SEX_BTSX' : 'both'})
```

```python
unique = row_mal['Gender'].unique()
print(unique)
```

```python
row_ob['CI_Width'] = row_ob['UpperBound'] - row_ob['LowerBound']
row_mal['CI_Width'] = row_mal['UpperBound'] - row_mal['LowerBound']
```

```python
def ob_level(val):
    if val >= 30:
        return 'high'
    elif 25 <= val <= 29.9:
        return 'moderate'
    else:
        return 'low'
```

```python
row_ob['obesity_level'] = row_ob['Mean_Estimate'].apply(ob_level)
```

```python
def mal_level(val):
    if val >= 20:
        return 'high'
    elif 10 <= val <= 19.9:
        return 'moderate'
    else:
        return 'low'
```

```python
row_mal['malnutrition_level'] = row_mal['Mean_Estimate'].apply(mal_level)
```

```python
special_cases = {
                    'GLOBAL': 'Global',
                    'WB_LMI': 'Low & Middle Income',
                    'WB_HI': 'High Income',
                    'WB_LI': 'Low Income',
                    'EMR': 'Eastern Mediterranean Region',
                    'EUR': 'Europe',
                    'AFR': 'Africa',
                    'SEAR': 'South-East Asia Region',
                    'WPR': 'Western Pacific Region',
                    'AMR': 'Americas Region',
                    'WB_UMI': 'Upper Middle Income'}
```

```python
import pycountry

def code_to_name(code):
    if code in special_cases:
        return special_cases[code]  

    country = pycountry.countries.get(alpha_3=code)  
    return country.name if country else code
```

```python
row_ob['Country'] = row_ob['Country'].apply(code_to_name)
```

```python
row_mal['Country'] = row_mal['Country'].apply(code_to_name)
```

```python
import pymysql

connection = pymysql.connect(
    host = 'localhost',
    user = 'root',
    password = 'root',
    database = 'Nutrition_Paradox'
)

cur = connection.cursor()
```

```python
cur.execute('''CREATE TABLE obesity_1 (
    region VARCHAR(100),
    gender VARCHAR(10),
    year INT,
    lower_bound FLOAT,
    upper_bound FLOAT,
    mean_estimate FLOAT,
    country VARCHAR(100),
    ci_width FLOAT,
    obesity_level VARCHAR(20)
);
''')
```

```python
cur.execute('''CREATE TABLE malnutrition (
    region VARCHAR(100),
    gender VARCHAR(10),
    year INT,
    lower_bound FLOAT,
    upper_bound FLOAT,
    mean_estimate FLOAT,
    country VARCHAR(100),
    ci_width FLOAT,
    malnutrition_level VARCHAR(20)
);
''')
```

```python
for index, row in row_ob.iterrows():
    cur.execute("""
        INSERT INTO obesity_1 (Region, Gender, Year, Lower_Bound, Upper_Bound, Mean_Estimate, Country, CI_Width, obesity_level)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, (row['Region'], row['Gender'], row['Year'], row['LowerBound'], row['UpperBound'], row['Mean_Estimate'], row['Country'], row['CI_Width'], row['obesity_level']))
connection.commit() 
```

```python
#1.Top 5 regions with the highest average obesity levels in the most recent year(2022)

cur.execute('''SELECT * 
FROM (
    SELECT 
        region, 
        AVG(mean_estimate) AS avg_ob_level, 
        DENSE_RANK() OVER (ORDER BY AVG(mean_estimate) DESC) AS rnk_ob_level
    FROM obesity_1
    WHERE year = 2022
    GROUP BY region
) AS ranked
WHERE rnk_ob_level <= 5;

''')
```
